```python
#importing dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
#importing and reading csv file
df = pd.read_csv("D:\zomato\zomato.csv")
df.head()
```

```python
df.columns
```

```python
df.drop(['url','book_table','online_order','menu_item','rest_type','listed_in(city)'],axis=1)
```

```python
#dropm duplicate values
df.drop_duplicates(inplace = True)
df.shape
```

```python
#cleaning the rate column
df['rate'].unique()
```

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#dropping null values
df.dropna(inplace = True)
df.head()
```

```python
#filling null values in the rate column with mean
df.rename(columns = {'approx_cost(for two people)':'cost2plates','listed_in(type)':'type'},inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['cost2plates'].unique()
```

```python
def handlecomma(value):
    value = str(value)
    if',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

df['cost2plates'] = df['cost2plates'].apply(handlecomma)
df['cost2plates'].unique()
    
```
